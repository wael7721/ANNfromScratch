## 2. Non-linear Activation

### Sigmoid
$ S(x) = e^x/(1+e^x)$

usually used in binary classification

```python
def sigmoid(x):
    return np.exp(x)/(np.exp(x)+1)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5)
y= sigmoid(x)
fig = plt.figure(figsize = (10, 5))
plt.axhline(y = 0,color="black", linestyle = '--')
# Create the plot
plt.plot(x, y)
 
# Show the plot
plt.show()
```

### TanH
$ tanH(x) = \frac {2}{(1+e^{-2x})}-1 $\
corrects issues of [Vanishing gradient problem](https://en.wikipedia.org/wiki/Vanishing_gradient_problem). "when the derivatives or slopes of the activation functions become progressively smaller as we move backward through the layers of a neural network."

```python
def tanH(k):
    return 2/(1+np.exp(-2*k))-1
```

```python
x = np.linspace(-5, 5)
y= tanH(x)
fig = plt.figure(figsize = (10, 5))
# Create the plot
plt.plot(x, y)
plt.axhline(y = 0,color="black", linestyle = '--')
 
# Show the plot
plt.show()
```

### RELU
Less computationally expensive\
More efficient for backpropagation

```python
def relu(x):
    return np.maximum(0,x)
```

```python
x = np.linspace(-5, 5)
y= relu(x)
fig = plt.figure(figsize = (10, 5))
# Create the plot
plt.plot(x, y)
plt.axhline(y = 0,color="black", linestyle = '--')
# Show the plot
plt.show()
```

### Leaky RELU
allows a small, non-zero gradient for negative inputs, preventing neurons from becoming inactive

```python
def leakyRelu(x,k=0.1):
    #choose k constant, else set to 0.1
    return np.maximum(k*x,x)
```

```python
x = np.linspace(-5, 5)
y= leakyRelu(x)
fig = plt.figure(figsize = (10, 5))
# Create the plot
plt.plot(x, y)
plt.axhline(y = 0,color="black", linestyle = '--')
# Show the plot
plt.show()
```

### Softmax
Multi Class classification
Normalizes the values to between 0 and 1

```python
def softMax(x):
    x_exp = np.exp(x - np.max(x))
    return x_exp / np.sum(x_exp,axis=0)
```
